  > **Pregunta**: El parámetro `uic` puede tener varios valores y cada uno significa una cosa. Detállalo usando un celda Markdown y consultando la [documentación de Ahkab](https://buildmedia.readthedocs.org/media/pdf/ahkab/latest/ahkab.pdf).

 > Parámetro UIC en un análisis .TRAN
<ol>
<li><b>uic=0</b></li>
Todos los voltajes e intensidades se inferirán como 0 en t = tstart. Es el valor por defecto si no se especifica otra cosa
<li><b>uic=1</b></li>
El status del circuito en t = tstart parte del resultado de un análisis .Op previo
<li><b>uic=2</b></li>
Igual que con el valor anterior solo que en el análisis .Op previo se habrán tenido en cuenta los valores de intensidad en bobinas y voltage en condensadores especificados en su parámetro 'ic'
<li><b>uic=3</b></li>
Obliga el uso de otro parámetro extra llamado IC_LABEL. Uic=3 obliga a que exista una directiva .ic (initial conditions) en otra parte del programa cuyo name coincida con el valor del parámetro IC_LABEL
</ol>

 > Directiva .IC/Atributo IC en bobinas y condensadores
<ol>
<li><b>Atributo Ic en bobinas</b></li>
Especifica el valor que éste elemento aportará en el caso de una análisis .TRAN con UIC=2
<li><b>Atributo Ic en condensadores</b></li>
Especifica el valor que éste elemento aportará en el caso de una análisis .TRAN con UIC=2
<li><b>Directiva .Ic</b></li>
En un análisis .TRAN con UIC=3 se buscará la directiva .ic coincidente IC_LABEL = .IC NAME. La directiva .Ic permite establecer condiciones iniciales para los nodos del circuito.
</ol>
