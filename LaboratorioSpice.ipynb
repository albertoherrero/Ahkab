  > **Pregunta 11:** comprueba que la ecuación anterior está ajustada a nivel dimensional, es decir, que la naturaleza de lo que está a ambos lados del signo igual es la misma. Realiza este ejercicio con LaTeX en una celda Markdown.

  Dado que la _ley de Ohm_ se puede expresar indistintamente:

  $$
  I = \frac{V}{R} \Leftrightarrow V = I \cdot R
  $$
  y La dimensión de _R_ (Resistencia es):
  $$
  \left[R\right]=M\: L^2\:I^{-2}\: T^{-3}
  $$
  y la de _V_ (Potencial):
  $$
  \left[V\right]=M\: L^2\:T^{-3}\:I^{-1}
  $$

  Por tanto si las dimensiones deben coincidir:

  $$
  \left[I\right] = \frac{\left[V\right]}{\left[R\right]}
  $$

  Despejando:
  $$
  [I] = \frac{M\: L^2\:T^{-3}\:I^{-1}}{M\: L^2\:I^{-2}\: T^{-3}}
  $$
  $$
  [I] = \frac{1}{I^{-1}}
  $$
  $$
  [I] = I
  $$

  Es decir, coinciden las dimensiones. Aunque _simpy_ dispone de un módulo _Dimension_ para tal efecto:
 https://docs.sympy.org/latest/modules/physics/units/philosophy.html#dimensions

  También, mediante _sympy_ podemos comprobar que sus unidades son las mismas,

  > **Pregunta 20**: ic=0.

 El parametro ic=0 en un netlist indica que el valor incial de un componente es cero. Principalmente se aplica a los condensadores condensadores. Es el último parametro en su definición.
 Con el comando .tran en un netlist se pretende analizar el comportamiento de un circuito en un periodo de tiempo dado entre el valor tinicial y final a intérvalos dados por tstep.


  > **Pregunta 21**: Carrera de condensadores.


```python
get_ipython().run_cell_magic('writefile', '"carrera de condensadores.ckt"', '* Carga condensador\nv1 0 1 type=vdc vdc=10\nvdd1 2 1 vdc=0 type=vdc\nvdd2 3 1 vdc=0 type=vdc\nc1 2 4 47u\nc2 3 4 22u\nr1 0 4 3.3k\n.op\n.tran tstep=0.01 tstart=11.8 tstop=12 uic=0\n.end\n')
circuito_y_análisis = ahkab.netlist_parser.parse_circuit("carrera de condensadores.ckt")
circuito = circuito_y_análisis[0]
análisis_en_netlist = circuito_y_análisis[1]
lista_de_análisis = ahkab.netlist_parser.parse_analysis(circuito, análisis_en_netlist)
lista_de_análisis[1]['outfile'] = "simulación tran.tsv"
resultados = ahkab.run(circuito, lista_de_análisis)
print(resultados['op'])
```

```python
# import matplotlib.pyplot as plt
figura = plt.figure()
plt.title("Carrera de condensadores")
plt.plot(resultados['tran']['T'], resultados['tran']['I(VDD1)'], label="C1")
plt.plot(resultados['tran']['T'], resultados['tran']['I(VDD2)'], label="C2")
plt.ylabel('Intensidad (A)')
plt.xlabel('Tiempo (s)')
plt.legend()
```
